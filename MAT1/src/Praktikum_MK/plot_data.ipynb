```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("messung1.csv")
df1["Zeit_s"] -= 105# Zeit in Sekunden ab 80s
df1 = df1[df1["Zeit_s"] >= 0]

t1 = df1["Zeit_s"]
T_ref1 = df1["T_ref_C"]
dig1 = df1["RTD_dig"]
T_Pt100_1 = df1["T_Pt100_C"]

df1.head()
df1.to_csv("messung1_korrigiert.csv", index=False)
```

```python
t1.iloc[-1]
T_ref1.iloc[-1]
```

```python
df2 = pd.read_csv("messung2.csv")
t2 = df2["Zeit_s"]
dig2 = df2["RTD_dig"]
T_Pt100_2 = df2["T_Pt100_C"]

df2.head()
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Linke Achse: Temperaturen
l1, = ax1.plot(t1, T_ref1, label="T_ref (DS18B20)")
l2, = ax1.plot(t1, T_Pt100_1, label="T_Pt100")
ax1.set_xlabel("Zeit [s]")
ax1.set_ylabel("Temperatur [C]")
ax1.grid(True)

# Rechte Achse: Digitalwert
ax2 = ax1.twinx()
l3, = ax2.plot(t1, dig1, label="RTD digital", color="green")
ax2.set_ylabel("u [digit]")

# Gemeinsame Legende
ax1.legend(handles=[l1, l2, l3], loc="best")

plt.title("Temperaturverlauf und digitaler RTD-Wert ber Zeit")
plt.tight_layout()
plt.savefig("v12A.png")
plt.show()
```

```python
plt.plot(T_ref1, dig1)
plt.title("RTD digitaler Wert in Abhngigkeit von T_ref")
plt.xlabel("T_ref [C]")
plt.ylabel("u [digit]")
plt.grid(True)
plt.savefig("v12A_2.png")
```

```python
plt.plot(t2, T_Pt100_2)
plt.title("Sprungversuch")
plt.xlabel("Zeit [s]")
plt.ylabel("Temperatur [C]")
plt.grid(True)
```
