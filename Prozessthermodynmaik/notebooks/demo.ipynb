### Beispielwerte

```python
p = 101325       # Pa
T = 20.0       # °C
ps = get_saturation_properties(T=T, return_value='ps')    # Pa
phi = 0.6        # relative Luftfeuchte [-]
x = eq45x(p, ps, phi)     # Wassergehalt
xs = eq45xs(p, ps)        # Sättigungs-Wassergehalt
v1 = 0.83                 # spezifisches Volumen trockene Luft (m³/kg)

print("Eingabedaten:")
print(f"p = {p} Pa, ps = {ps:.1f} Pa, phi = {phi}, T = {T_C} °C")
print(f"x  = {x:.6f} kgW/kgL, xs = {xs:.6f} kgW/kgL")
print("-"*80)
```

### (4.1) x

```python
mw = 0.008     # kg Wasser
ml = 1.0       # kg Luft
x_41 = eq41x(mw, ml)
print(f"(4.1) x = {x_41:.6f} kgW/kgL")
```

### (4.5) x, φ, xs, ps 

```python
x = eq45x(p, ps, phi) # kgW/kgL
phi = eq45phi(p, ps, x)
ps = eq45ps(p, x, phi) # Pa
xs = eq45xs(p, ps)     # kgW/kgL
```

### (4.8) Spezifisches Volumen – ungesättigte / gesättigte Luft

```python
v = eq48v(p, T, x)      # m³/kg
```

### (4.9) Spezifisches Volumen – übersättigte Luft

```python
v1 = get_saturation_properties(T=T, return_value="v'")
v = eq49v(p, T, x, xs, v1)  # m³/kg
```

### (4.10) Enthalpie und Temperatur bei ungesättigter Luft

```python
h = eq410h(T, x)     # kJ/kgL
T = eq410T(h, x)    # °C
```

### (4.11) Enthalpie und Temperatur bei übersättigter Luft, T > 0 °C

```python
h = eq411h(T, x, xs)    # kJ/kgL
T = eq411T(h, x, xs)  # °C
```

### (4.12) Enthalpie und Temperatur bei übersättigter Luft, T < 0 °C
