```python
import numpy as np
import time
import scipy.optimize as opt
```

```python
a = 0.956
b = 2.12e-4
c = 2.51e-8
d = 2.58e-11
R = 0.287

T1 = 20+273.15
p1 = 0.1e6
p2 = 1.5e6
```

```python
term1 = R * np.log(p2 / p1)

def residual(T2):
    term2 = a * np.log(T2 / T1)
    term3 = b * (T2 - T1)
    term4 = c/2 * (T2**2 - T1**2)
    term5 = d/3 * (T2**3 - T1**3)
    res = term1 - (term2 + term3 + term4 + term5)
    return res

# Ableitung
def residual_prime(T2):
    return -(a / T2 + b + c*T2 + d*T2**2)

# Startwert
T2 = 600.0  # z.B. 300 K

for i in range(30):  # max. 30 Iterationen
    f = residual(T2)
    fp = residual_prime(T2)
    T2_new = T2 - f/fp
    print(f"Iter {i}: T2 = {T2_new:.6f} K, f = {f:.6e}")
    if abs(T2_new - T2) < 1e-10:
        break
    T2 = T2_new

print("\nFinaler Wert:", T2, "K")
```
