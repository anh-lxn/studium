```python
import sys, os
project_root = os.path.abspath(os.path.join(os.getcwd(), ".."))
sys.path.append(os.path.join(project_root, "src"))
%load_ext autoreload
%autoreload 2
from coolprop_utils import get_saturation_properties
from FL import *
```

```python
p = 100000 # Pa
```

a)

```python
T = 26 # °C
phi = 0.35
ps = get_saturation_properties(T=T, return_value="ps") # Pa
x = eq45x(p, phi, ps)
xs = eq45xs(p, ps)
if x < xs:
    print("ungesättigte/gesättigte Luft")
elif x >= xs:
    print("übersättigte Luft")

v = eq48v(p, T+273.15, x) # m³/kg
h = eq410h(T, x)

print(f"Wassergehalt x: {x*1000} gW/kgL")
print(f"Spezifisches Volumen v: {v} m³/kg")
print(f"Spezifische Enthalpie h: {h} kJ/kg")
```

b)

```python
x = 13.113 / 1000 # kgW/kgL
phi = 1
ps = eq45ps(p, x, phi)
T = get_saturation_properties(p=ps, return_value="T")
v = eq48v(p, T+273.15, x)
h = eq410h(T, x)

print(f"Temperatur T: {T} °C")
print(f"relative Luftfeuchte φ: {phi*100} %")
print(f"Spezifisches Volumen v: {v} m³/kg")
print(f"Spezifische Enthalpie h: {h} kJ/kg")
```

```python
T = 43 # °C
x = 60 / 1000 # kgW/kgL
ps = get_saturation_properties(T=T, return_value="ps") # Pa
phi = eq45phi(p ,ps , x)
# Phi ist größer als 1 -> übersättigte Luft
phi = None
xs = eq45xs(p, ps)
v1 = get_saturation_properties(T=T, return_value="v'") # m³/kg
v = eq49v(p, T+273.15, x, xs, v1)
h = eq411h(T, x, xs)

print(f"relative Luftfeuchte φ: nicht definiert")
print(f"Spezifisches Volumen v: {v} m³/kg")
print(f"Spezifische Enthalpie h: {h} kJ/kg")
```
