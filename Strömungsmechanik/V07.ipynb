# Technische Strömungsmechanik
## Nachbereitungsfragen – Vorlesung 07 : Potentialströmungen

1. Was unterscheidet eine partielle von einer gewöhnlichen Differentialgleichung (DGL)?
- gewönhnliche DGL: Ableitungen nach einer unabhängigen Variable (z.B. Zeit)
- partielle DGL: Ableitungen nach mehreren unabhängigen Variablen (z.B. Raumkoordinaten)

2. Nennen Sie 3 Typen von partiellen DGL.
Was beschreibt der Typ einer partiellen DGL?

- hyperbolisch -> Wellenausbreitung
- parabolisch -> Diffusion
- elliptisch -> Wärmeleitung
- der Typ beschreibt Transport von Informationen

3. Welchen Typ besitzt die Laplace-Gleichung?

- elliptisch

4. Welche Angaben beschreiben ein DGL-Problem vollständig?

- die DGL selbst
- Intergrationsgebiet (Gebiet in dem die DGL gelöst werden soll)
- Randbedingungen am Rand des Integrationsgebiets
- Falls zeitabhängig: Anfangsbedingungen zum Startzeitpunkt

5. Können die Randbedingungen eines DGL-Problems beliebig gewählt werden?

- Nein
- Müssen so gewählt werden:
  1. Das Problem überhaupt eine Lösung besitzt
  2. Die Lösung eindeutig ist
  3. Der Typ der PDGL berücksichtigt wird
  4. Physikalisch sinnvoll sind


6. Was ist eine Neumann-Randbedingung?
Wann wird diese als homogen bezeichnet?

- Für ein Potiential $\phi$:
$$
\frac{\partial \phi}{\partial n} = g
$$

- homogen wenn g = 0 ist

7. Welche Arten von Randbedingungen sind für Potentialströmungen geeignet?
Notieren Sie jeweils die Bedingung für die Geschwindigkeit und das Geschwindigkeitspotential.

- Neumann-Randbedingung (Normalgeschwindgikeit gegeben):
  - Geschwindigkeit: $ \underline{u} \cdot \underline{n} = u_n = g $
  - da $ \underline{u} = \nabla \phi$, gilt $ \frac{\partial \phi}{\partial n} = g $
- Dirichlet-Randbedingung (Geschwindigkeitspotential gegeben):
  - Geschwindigkeit: $ \phi = \phi_0 $ auf dem Rand


8. Wenn $\phi$ eine Potentialströmung beschreibt, beschreibt dann Φ + C, C ∈ ℝ, die gleiche Strömung? Begründen Sie.

- Ja, da die Ableitung einer Konstante 0 ist:
$$
\underline{u} = \nabla \phi = \nabla (\phi + C)
$$

9. Wann ist ein Gebiet einfach zusammenhängend?

- Ein Gebiet ist einfach zusammenhängend, wenn jede geschlossene Kurve darin zu einem Punkt zusammengezogen werden kann, ohne das Gebiet zu verlassen.
- Beispiele:
  - einfach zusammenhängend: Kreis, Rechteck, Kugel
  - nicht einfach zusammenhängend: Ring, Donut

10. Gegeben ist ein Schweizer Käse.
Ist das Gebiet, das der Käse ohne die Löcher einnimmt, einfach zusammenhängend?
Sie schneiden eine Scheibe mit Löchern davon ab.
Ist das Gebiet des Käses dann einfach zusammenhängend?

- Käse ohne Löcher -> einfach zusammenhängend
- Käse mit Löchern -> nicht einfach zusammenhängend

11. Welche zusätzliche Angabe ist bei nicht einfach zusammenhängenden Gebieten und den Randbedingungen für eine Potentialströmung nötig, um eine eindeutige Lösung für $\phi$ (abgesehen von einer frei wählbaren Konstanten) zu erhalten?

- Man muss die Zirkulation (oder Umlaufintegral der Geschwindigkeit) um jedes Loch / jede nicht zusammenhängende Komponente vorgeben.
- Warum? -> in einem Gebiet mit Löchern kann das Potential:
  - unterschiedlich "um das Lopch herum" verlaufen
  - mehrere gültige Potentiale erzeugen, die die gleiche Randbedingung erfüllen 

12. Gilt die Laplace-Gleichung für das Potential auch im Fall einer instationären Strömung?

- Ja ist egal
$$
\nabla^2 \phi = \nabla^2 \phi(x,y,z,t) = 0
$$

13. Unter welchen Voraussetzungen erhält man eine instationäre Lösung für die Laplace-Gleichung?

- Die Laplace-Gleichung selbst ist zeitunabhängig.
- Damit es instationär wird, muss sich was am Außenrand ändern
  - Form des Gebiets ändert sich mit der Zeit
  - Randbedingungen ändern sich mit der Zeit (z.B: bewegte Wände)


14. Gegeben sei das Stromlinienbild einer Strömung.
Kann man aus diesem Bild den absoluten Betrag der Strömungsgeschwindigkeit ablesen? Begründen Sie.

- Nein, da Stromlinien nur die Richtung der Strömung angeben.
- Der Abstand zwischen den Stromlinien gibt nur relative Geschwindigkeitsunterschiede an

15. Was bedeutet es, wenn man sagt: „die Volumenkraft hat ein Potential“?

$$
\underline{f} = - \nabla G
$$

- das beudetet die Kraft ist konservativ
- die Arbeit hängt nicht vom Weg ab, sondern nur von Anfangs- und Endpunkt
- Beispiele: Gravitationskraft, elektrostatische Kraft, Druckkräfte in Potentialströmungen


16. Beschreiben Sie die zwei Wege, ein Potentialströmungsproblem zu lösen.

- Direkte Lösung der Laplace-Gleichung mit Randbedingungen
- Superposition von Grundlösungen
  - Man setzt die Strömung aus bekannten Potentialelementen zusammen (Quelle, Senke, Doppeltsrömung, Wirbel)
  - $\phi = \phi_1 + \phi_2 + \dots + \phi_n$

17. Wo können in einer Potentialströmung Wände eingefügt werden, ohne die Strömung zu beeinflussen?
Ist das auch für reibungsbehaftete Strömungen möglich?

- Überall dort wo die Strömung tangential zur Wand verläuft (also keine Normalgeschwindigkeit an der Wand)
- $u_n = 0 \text{ somit gilt } \frac{\partial \phi}{\partial n} = 0$



18. Umströmte Körper werden oft „stromlinienförmig“ gestaltet.
Erklären Sie den Ausdruck im Kontext der Potentialtheorie.

- stromlinienförmig: Die Form des Körpers folgt den Stromlinien einer idealen (Potential-)Strömung, sodass die Strömung ohne Ablösung und mit minimaler Störung um den Körper herumgeführt wird.


19. Welche Potentialströmungen müssen kombiniert werden, um die Strömung um den Rankine-Halbkörper zu erzeugen?
Skizzieren Sie die Stromlinien.
Wie viele Staupunkte gibt es in der Strömung?

- Rankine-Halbkörper entsteht durch die Superposition einer Quelle und einer Gleichströmung mit Geschwindigkeit $U_\infty$ in x-Richtung.
- Warum?
  - Gleichströmung drück Partikel nach rechts
  - Quelle bläst Partikel nach außen

20. Wie verändert sich der Rankine-Halbkörper, wenn die Stärke der Quelle erhöht wird?

- Je größer die Quellstärke $Q$, desto mehr Fluid wird nach außen gedrückt.
- Dadruch wird Rankine länger (in Strömungsrichtung) und dicker (durchmesser).

22. Schreiben Sie den Konvektionsterm der Navier–Stokes-Gleichungen so, dass die Wirbelstärke auftaucht.

- Konvektionsterm: $\quad (\underline{u} \cdot \nabla) \underline{u}$
- Wirbelstärke: $\quad \underline{\omega} = \nabla \times \underline{u}$
- Vektoridentität: 
$$
(\underline{u} \cdot \nabla) \underline{u} = \nabla \left( \frac{1}{2} ||\underline{u}||^2 \right) - \underline{u} \times (\nabla \times \underline{u})
$$
- Somit:
$$
\boxed{
(\underline{u} \cdot \nabla) \underline{u} = \nabla \left( \frac{1}{2} ||\underline{u}||^2 \right) - \underline{u} \times \underline{\omega}
}
$$

23. Schreiben Sie den Reibungsterm der Navier–Stokes-Gleichungen so, dass die Wirbelstärke auftaucht.  

- Reibungsterm: $\quad \nu \nabla^2 \underline{u}$
- Vektoridentität: $\nabla^2 \underline{u} = \nabla (\nabla \cdot \underline{u}) - \nabla \times (\nabla \times \underline{u})$
- inkompressible Strömung: $\nabla \cdot \underline{u} = 0$
- Somit: 
$$
\boxed{
\nu \nabla^2 \underline{u} = - \nu \nabla \times \underline{\omega}
}
$$
