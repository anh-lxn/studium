2. Wann wird eine komplexe Funktion harmonisch genannt?

- eine Funktion $f(z)$ ist harmonisch, wenn Real- und Imaginärteil die Laplace-Gleichung erfüllt:
$$\nabla^2 f_R = \frac{\partial^2 f_R}{\partial x^2} + \frac{\partial^2 f_R}{\partial y^2} = 0 \quad \text{und} \quad \nabla^2 f_I = \frac{\partial^2 f_I}{\partial x^2} + \frac{\partial^2 f_I}{\partial y^2} = 0$$

5. Gibt es komplexe analytische Funktionen, die keine Potentialströmung darstellen? Gibt es Potentialströmungen, die sich nicht als komplexe analytische Funktion schreiben lassen? Begründen Sie.

- ja, es gibt komplexe analytische Funktionen, die keine Potentialströmung darstellen: Sie muss eine physikalisch sinvolle Strömungs beschreiben (manche analytische Funtkionen liefern keine reelen Geschwindigkeitsfelder)
- nein, jede 2D Potentialströmung lässt sich als komplexe analytische Funktion schreiben, weil jede 2D, inkompressible, rotaionsfreie Strömung einn Geschwidigspotential $\Phi$ und eine Stromfunktion $\Psi$ besitzt $$\Rightarrow f(z) = \Phi + i \Psi$$

9. Welcher Zusammenhang besteht zwischen der Quellstärke $Q$ und dem auf die Breite bezogenen Volumenstrom $\dot{V}/b$, der aus der Quelle entspringt?

$$
Q = \frac{\dot{V}}{b}
$$

- $Q$: Quellstärke [m²/s]
- $\dot{V}/b$: Volumenstrom [m²/s]
- $b$: Breite in z-Richtung [m]

10. Welche Ähnlichkeit besteht zwischen Quellströmung und Potentialwirbel?

- gleiche mathematische Form der komplexen Potentialfunktion:
  - Quellströmung: $f(z) = \frac{Q}{2\pi} \ln z$
  - Potentialwirbel: $f(z) = -i \frac{\Gamma}{2\pi} \ln z$
- gleiche Geschwindigkeitsabhängigkeit vom Radius: $v \sim \frac{1}{r}$
- Singularität im Ursprung $r=0$
- Strom- und Potentiallinien vertauscht
  - Quelle:
    - SL: radial
    - PL: Kreise
  - Wirbel:
    - SL: Kreise
    - PL: radial

11. Wie groß ist die Geschwindigkeit im Ursprung der Quellströmung?

- unendlich groß, da $v_r = \frac{Q}{2 \pi r}$ und für $r \to 0$ gilt $v_r \to \infty$

17. Leiten Sie die komplexe Potentialfunktion der Keilströmung nach $z$. Welche Größe erhalten Sie dann?

- Keilströmung: $f(z) = Cz^n$
- Ableitung: $\frac{df}{dz} = C n z^{n-1} = u - iv \rightarrow \text{liefert komplexe Geschwindigkeit}$

![Keilströmung](img/v09_17.png)

18. Wie hängt der Öffnungswinkel einer Keilströmung vom Exponenten in der komplexen Potentialfunktion ab?

- in Polarform: $f(z) = C r^n e^{i n \theta}$
- folgender Zusammenhang gilt $\alpha = \frac{\pi}{n}$

19. In welche Richtung verläuft eine Keilströmung auf der positiven x-Achse? Wie verläuft $q$ entlang der positiven x-Achse?

- $u = Cnx^{n-1}$ und $v = 0$
- je nach Vorzeichen von $C$ verläuft die Strömung in positive oder negative x-Richtung
- $q = \sqrt{u^2 + v^2} = |Cn| x^{n-1}$ also $q \sim x^{n-1}$
- Interpretation:
  - für $n>1$ wächst die Geschwindigkeit mit zunehmendem x
  - für $n<1$ fällt die Geschwindigkeit mit zunehmendem x
  - für $n=1$ ist die Geschwindigkeit konstant (Gleichströmung)

20.  Skizzieren Sie eine Keilströmung mit einem Öffnungswinkel von 135°. Wie lautet die komplexe Potentialfunktion, wenn diese Strömung so um den Koordinatenursprung gedreht ist, dass die Wand auf der Zuströmseite auf der negativen x-Achse liegt?

- $\alpha = \frac{180°}{n} \Rightarrow n = \frac{180°}{135°} = \frac{4}{3} \rightarrow f(z) = C z^{\frac{4}{3}}$
- Drehung um 180°:
  - $f(z) = C e^{i \pi} z^{\frac{4}{3}} = -C z^{\frac{4}{3}}$

21.  Hat die Keilströmung eine Singularität bei $z=0$?

- Fallunterscheidung für $q \sim x^{n-1}$:
  - $n>1$: nein, da $q \to 0$
  - $n=1$: nein, da $q = \text{const.}$
  - $n<1$: ja, da $q \to \infty$

22. Ein horizontaler Freistrahl trifft auf eine schräg angestellte Platte. Wie lässt sich diese Strömung mithilfe einer Keilströmung darstellen?

- Öffnungswinkel/Schrägstellungswinkel $\alpha$ messen
- Exponent $n$ berechnen: $n = \frac{\pi}{\alpha}$
- Keilströmung $f(z) = C z^n$ aufstellen

24. Wie berechnet man das Ringintegral einer analytischen Funktion entlang einer beliebigen geschlossenen Kurve?

$$
\oint_{\gamma} f(z) dz = 0 \quad \text{wenn im Gebiet keine Singularität vorliegt }
$$

25. Wie lautet die komplexe Zirkulation einer Potentialströmung, wenn innerhalb der Integrationskurve Singularitäten vorkommen?

- geschlossenes Wegintegral ist nicht mehr null, sondern:
$$
\oint_{\gamma} f(z) dz = 2 \pi i \sum \text{Res}(f, \gamma)
$$

- für eine Laurant-Reihe gilt:
$$
\text{Res }f(z) = c_{-1}
$$

26. Wie hängt die komplexe Zirkulation $\Gamma_C$ mit der reellen Zirkulation $\Gamma$ zusammen, wie mit der Quellstärke?

$$
\Gamma_C = \Gamma + i Q = \underbrace{\oint_{\gamma} udx + vdy}_{\text{reelle Zirkulation}} + i \underbrace{\oint_{\gamma} (udy - vdx)}_{\text{Quellstärke}}
$$

27. Weisen Sie nach / Berechnen Sie das Ringintegral entlang einer Kreislinie mit dem Radius $\sqrt{2}$ um den Ursprung für $f(z) = z + \frac{1}{z}$. Berechnen Sie dasselbe für $\tilde{f}(z) = z^2 + \frac{1}{z^2}$.

- $f(z) = z + \frac{1}{z}$:
$$
\oint_C f(z) dz = \oint_C \left( z + \frac{1}{z} \right) dz = \oint_{\gamma} z dz + \oint_{\gamma} \frac{1}{z} dz = 0 + 2 \pi i = 2 \pi i \rightarrow \text{Res } \frac{1}{z} = 1
$$

- $\tilde{f}(z) = z^2 + \frac{1}{z^2}$:
$$
\oint_C \tilde{f}(z) dz = \oint_C \left( z^2 + \frac{1}{z^2} \right) dz = \oint_{\gamma} z^2 dz + \oint_{\gamma} \frac{1}{z^2} dz = 0 + 0 = 0 \rightarrow \text{Res } \frac{1}{z^2} = 0
$$

28. Weisen Sie nach, dass das Integral $\oint_S (udy - vdx)$ die Summe der Quellstärken der eingeschlossenen Quellen darstellt.

- Normalvolumenstrom durch geschlossene Kurve $S$:
$$
\dot{V} = \oint_S (udy - vdx) = \oint_S \underline{u} \cdot \underline{n} ds
$$

- "$udy$" ist die Gehschwindigkeit $u$ in x-Richtung mal der Länge $dy$
- "$-vdx$" ist die Geschwindigkeit $v$ in y-Richtung mal der Länge $dx$
- misst also den Nettofluss nach außen und ist deshalb gleich der Summe der Quellstärken der eingeschlossenen Quellen

$$
\oint_S (udy - vdx) = \sum_i Q_i
$$


29. Berechnen Sie $\oint_S (udy - vdx)$ für eine Dipolströmung im Ursprung und $S$ gleich einem Kreis um den Ursprung mit dem Radius 1

- Dipol besteht Aus Quelle und Senke gleicher Stärke $Q$, also gilt $\sum_i Q_i = Q + (-Q) = 0$
- somit ist auch das Integral null:

$$
\oint_S (udy - vdx) = 0
$$
