# Technische Strömungsmechanik
## Nachbereitungsfragen – Vorlesung 06: Wirbelströmungen/Potentialströmungen

1. Was besagt der 3. Helmholtzsche Wirbelsatz?
Für eine Wirbelröhre gilt, dass die Zirkulation entlang einer geschlossenen Kurve, die die Wirbelröhre umschließt, konstant bleibt.
$$\Gamma_1 = \Gamma_2 = \oint_C \vec{u} \cdot d\vec{s}$$

2. Was lässt sich daraus für das zeitliche Verhalten einer Wirbelröhre ableiten?
Welcher Wirbelsatz muss dazu in Betracht gezogen werden?

- Wenn sich die Wirbelröhre bewegt, ändert sich ihre Form, aber die Zirkulation bleibt konstant.
- Der 2. Helmholtzsche Wirbelsatz besagt, dass sich Wirbellinien mit dem Fluid bewegen.
- Daraus folgt dass die Wirbelröhre mit der Strömung wandert.

3. Der Radius einer Wirbelröhre mit Kreisquerschnitt verdoppelt sich.
Wie verhält sich die Wirbelstärke im Inneren?

$$
\Gamma = \omega \cdot A = const.,  \quad A = \pi r^2, \quad r_2 = 2 r_1
$$
$$
\omega_1 \cdot \pi r_1^2 = \omega_2 \cdot \pi (2 r_1)^2,
$$
$$
\Rightarrow \quad \omega_2 = \frac{1}{4}\omega_1
$$
- Die Wirbelstärke nimmt um ein Viertel ab.


4. Wie verhalten sich Querschnitt und Wirbelstärke in einer Wirbelröhre, wenn diese in Längsrichtung gestreckt wird?

- Wenn die Röhre gestreckt wird, dann nimmt der Querschnitt ab.
- Da die Zirkulation konstant bleibt, muss die Wirbelstärke zunehmen.

5. Wo können Wirbellinien bei reibungsfreier Strömung enden?
- Bei reibungsfreier Strömung können Wirbellinien nur an festen Wänden oder im Unendlichen enden.

## Start Potentialströmungen


1. Mit welcher Kennzahl lässt sich die Bedeutung von Reibungseffekten in einer Strömung bewerten?
- Mit der Reynolds-Zahl $$ Re = \frac{\rho U L}{\mu} $$ lässt sich die Bedeutung von Reibungseffekten in einer Strömung bewerten
- Diese vergleich Trägheitskräfte zu viskosen Kräften (Reibungskräften).
- Re >> 1: Trägheitskräfte dominieren, Re << 1: Reibungskräfte dominieren

2. Wodurch ist das Grenzschichtkonzept charakterisiert?

- sehr dünner Bereich direkt an der Wand bei dem Reibung eine große Rolle spielt
- außerhalb der Grenzschicht kann die Strömung als reibungsfrei betrachtet werden
- An der Wand gilt die Haftbedingung: $ \underline{u} = 0 \Rightarrow $ Daraus folgt eine starker Geschwindigkeitsgradient in Wandnähe

3. Notieren Sie Charakteristika von Außenströmung und Grenzschichtströmung.

- Außenströmung:
  - Reibungsfrei
  - $\omega = \nabla \times \underline{u} = 0$
  - Druckfeld bestimmt die Strömung außen
- Grenzschichtströmung:
  - Reibungsbehaftet
  - $\omega = \nabla \times \underline{u} \neq 0$
  - dünne Schwerströmung durch Wandreibung entsteht

9. Können in einer rotationsfreien Strömung Geschwindigkeitsgradienten auftreten?

- rotationsfreie Strömung: $\omega = \nabla \times \underline{u} = 0$
  - bedeutet das es lokal keine Drehung des Fluidelements gibt
  - keine Wirbelstärke
- **ABER:** z.B. eine Potentialströmung um einen Zylinder 
  - über der Kugel nimmt die Geschwindigkeit zu
  - Trotzdem Strömung rotationsfrei
  - Geschwindigkeitsänderungen sind symmetrisch und führen zu keiner lokalen Rotation des Fluidelements


10. Wann kann die Existenz einer Grenzschicht vernachlässigt werden?

- Wenn die Reynolds-Zahl sehr groß ist
- Dadurch ist Grenzschicht sehr dünn
- Reibungseffekte sind vernachlässigbar
- man kann wie in der Außenströmung eine reibungsfreie Strömung annehmen mit $\omega = 0$

11. Wie verändert sich bei Vernachlässigung der Grenzschicht die Randbedingung an der Wand?

- An der Wann gitl dann die Gleitbedingung 
- $\underline{u} \cdot \underline{n} = u_n = 0$
- $u_t$ darf beliebig groß sein $\Rightarrow$ Strömung kann "schleifen"

12. Wie ist das Geschwindigkeitspotential definiert?
Was folgt daraus für die Wirbelstärke in einer Potentialströmung?

- Geschwindigkeitspotential $\phi$:

$$\underline{u} = \nabla \phi , \quad \phi \in \mathbb{R}$$

- Daraus folgt Wirbelstärke gleich 0:
$$
\omega = \nabla \times \underline{u} = \nabla \times (\nabla \phi) = 0
$$

13. Wie lautet die Laplace-Gleichung des Geschwindigkeitspotentials?

- Laplace-Gleichung:
$$
\nabla^2 \phi = \frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} + \frac{\partial^2 \phi}{\partial z^2} = 0
$$

14. Wie heißen Lösungen der Laplace-Gleichung?

- Lösungen der Laplace-Gleichung heißen Harmonische Funktionen.
- Beispiele:
  - $\phi(x,y) = ax + by + c$
  - $\phi = \ln(r)$
  - $\phi = \frac{1}{r}$

15. Welche Eigenschaft muss eine Potentialströmung erfüllen, damit die Laplace-Gleichung für das Geschwindigkeitspotential gilt?

- Die Strömung muss inkompressibel sein: $\nabla \cdot \underline{u} = 0$

16. Wie lautet die Helmholtz-Zerlegung für eine Potentialströmung?

$$
\underline {u} = \underline{u}^{(\omega)} + \underline{u}^{(\phi)} = 0 + \nabla \phi
$$


17. Was besagt das Superpositionsprinzip? Welche mathematische Eigenschaft wird dabei ausgenutzt?

- Linearkombinationen der Lösung ist wieder eine Lösung der Laplace-Gleichung.
- Grund ist dass Nabla ein linearer Operator ist.

$$
\nabla^2 (a\phi_1 + b\phi_2) = a\nabla^2 \phi_1 + b\nabla^2 \phi_2 = 0 + 0 = 0
$$
