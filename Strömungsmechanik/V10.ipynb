3. Was ist eine Laurent-Reihe?

- Reihenentwicklung einer komplexen Funktion, die auch negative Potenzen von $(z-a)$ enthält und zur Beschreibung von Singularitäten verwendet wird.
$$
f(z) = \underbrace{\sum_{n=0}^{\infty} c_n (z-a)^n}_\text{regulärer Teil (ohne Singularitäten)}  + \underbrace{\sum_{n=1}^{\infty} \frac{c_n}{(z-a)^n}}_\text{singulärer Teil (mit Singularitäten)}
$$

- Beispiel:
$$
f(z) = \underbrace{z}_\text{regulärer Teil} + \underbrace{\frac{1}{z}}_\text{singulärer Teil}
$$

4. Was ist eine vierfache Polstelle? Notieren Sie beispielhaft vier Funktionen mit einer vierfachen Polstelle.

- viefache Polstelle bei $z=a$, wenn die Funktion $f(z)$ in der Form
$$
f(z) = \frac{c_{-4}}{(z-a)^4}
$$

- Beispiel
  - $f(z) = \frac{1}{z^4}$
  - $f(z) = \frac{3}{z^4}$
  - $f(z) = \frac{1+z}{z^4}$
  - $f(z) = \frac{e^z}{z^4}$

5. Notieren Sie drei analytische Funktionen ohne Polstelle.

- analytisch $\Rightarrow$ komplex differenzierbar
- Beispiele:
  - $f(z) = z^2 + 2z + 1$
  - $f(z) = e^z$
  - $f(z) = \sin(z)$

6. Hat das Quadrat einer analytischen Funktion mit n-facher Polstelle auch eine n-fache Polstelle?

- nein, das Quadrat einer analytischen Funktion mit n-facher Polstelle hat eine 2n-fache Polstelle
$$
f(z) = \frac{1}{(z-a)^n} \Rightarrow f(z)^2 = \frac{1}{(z-a)^{2n}}
$$
- Beispiel: $f(z) = \frac{1}{(z-1)^3} \Rightarrow f(z)^2 = \frac{1}{(z-1)^6}$ 
- dreifache Polstelle wird zu sechsfacher Polstelle

8. Wie ist das Residuum einer analytischen Funktion $f$ in einem Punkt $a \in \mathbb{C}$ definiert?

$$
Res(f, a) = c_{-1}
$$
Oder als Integral:
$$
Res(f, a) = \frac{1}{2 \pi i} \oint_{C} f(z) \, dz
$$
wobei $C$ eine geschlossene Kurve um die Polstelle $a$ ist.


9. Was ist die Voraussetzung und was ist die Kernaussage des Residuensatzes?

- Voraussetzung: $f(z)$ ist anayltisch
- Kernaussage: Das Integral einer komplexen Funktion $f(z)$ entlang einer geschlossenen Kurve $C$ ist gleich der Summe der Residuen der Funktion innerhalb der Kurve, multipliziert mit $2 \pi i$.
$$
\oint_{C} f(z) \, dz = 2 \pi i \sum_k Res(f, a_k)
$$

12. Wozu dient die Singularitätenmethode? Erläutern Sie das strategische Vorgehen.

- Singularitätenmethode dient dazu, Potentialströmungen um feste Körper zu berechnen, indem der Körper durch eine geeignete Überlagerung von Elementarströmungen (Singularitäten) ersetzt wird.
- Strategie:
  - Baue Potentialströmungen mit Singularitäten zusammen
  - Mache das so, dass eine geschlossene (Teil-)Stromlinie entsteht, mit den Singularitäten innerhalb
  - Diese Stromlinie ist dann Körperkontur
  - Zirkulation entlang dieser Körperkontur aus Residuensatz mit Singularitäten innerhalb.
