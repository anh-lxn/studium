ausschreiben und anschließend das Kreuzprodukt ausf¨uhren und vereinfachen.
   $$
   \nabla \phi = \left( \frac{\partial \Phi}{\partial x}, \frac{\partial \Phi}{\partial y}, \frac{\partial \Phi}{\partial z} \right)
   $$

   $$
   \nabla \times \nabla \Phi = \left(
   \frac{\partial}{\partial y} \frac{\partial \Phi}{\partial z} - \frac{\partial}{\partial z} \frac{\partial \Phi}{\partial y},
   \frac{\partial}{\partial z} \frac{\partial \Phi}{\partial x} - \frac{\partial}{\partial x} \frac{\partial \Phi}{\partial z},
   \frac{\partial}{\partial x} \frac{\partial \Phi}{\partial y} - \frac{\partial}{\partial y} \frac{\partial \Phi}{\partial x}

   $$
   \frac{\partial^2 \Phi}{\partial y \partial z} = \frac{\partial^2 \Phi}{\partial z \partial y}, \quad
   \frac{\partial^2 \Phi}{\partial z \partial x} = \frac{\partial^2 \Phi}{\partial x \partial z}, \quad
   \frac{\partial^2 \Phi}{\partial x \partial y} = \frac{\partial^2 \Phi}{\partial y \partial x}
   $$
   
   Damit folgt 
   $$
   \nabla \times \nabla \Phi = (0, 0, 0)
   $$

5. Wie kann auf mathematischem Wege ein rotationsfreies Geschwindigkeitsfeld u erzeugt werden? Was bedeuten die Begriffe „Potentialfunktion“ und „Potentialfeld“?
Ein Feld ist rotationsfrei, wenn gilt:
$$
\nabla \times \underline{u} = 0
$$
Mathematisch erzeigt man ein solches Feld, indem man $\underline{u} $ als Gradient eines Skalarpotentials (-feldes) $\phi$ definiert:
$$
\underline{u} = \nabla \phi
$$

Dann gilt auch automatisch:
$$
\nabla \times \underline{u} = \nabla \times \nabla \phi = 0
$$  

- **Potentialfunktion**: Skalarfeld $\phi$, dessen Gradient das Geschwindigkeitsfeld $\underline{u}$ ergibt.
- **Potentialfeld**: Geschwindigkeitsfeld $\underline{u}$, das als Gradient eines Skalarfeldes $\phi$ dargestellt werden kann.

8. Wann kann eine Str¨omung als Potentialstr¨omung bezeichnet werden? Nutzen Sie in

Eine Strömung kann als Potentialströmung bezeichnet werden, wenn das Geschwindigkeitsfeld $\underline{u}$ wirbelfrei ist, also wenn der Wirbelanteil der Helmholtz-Zerlegung verschwindet:
$$
\underline u = \underbrace{\underline u^{(w)}}_{\text{Wirbelanteil}} + \underbrace{\underline u^{(\phi)}}_{\text{Potentialanteil}}
$$  

$$

$$

Somit gilt:
$$
\underline{u} = \underline u^{(\phi)} = \nabla \phi
$$

21. Warum bewegt sich ein Wirbelring in einer ansonsten ruhigen Umgebung?

Ein Wirbelring bewegt sich, weil seine eigene Wirbelstärke über das Biot–Savart-Gesetz ein Geschwindigkeitsfeld erzeugt, das an der Lage des Rings eine axiale Geschwindigkeit ergibt.
Dadurch „zieht“ sich der Ring selbst in eine Richtung voran – ganz ohne äußere Strömung.
