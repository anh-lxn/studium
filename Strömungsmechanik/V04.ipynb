5. Wie l¨asst sich die von einer unendlich ausgedehnten Wirbelschicht im Punkt P induzierte Geschwindigkeit berechnen?
$$
\underline{u}^{(oben)} - \underline{u}^{(unten)} = \underline{\gamma}
$$  
- $\underline{\gamma}$ ... Wirbelstärke pro Flächeneinheit 
- $\underline{u}^{(oben)}$ ... Geschwindigkeit oberhalb der Wirbelschicht
- $\underline{u}^{(unten)}$ ... Geschwindigkeit unterhalb der Wirbelschicht

$$
\text{Oberhalb der Schicht: } \underline{u}^{(oben)} = +\frac{1}{2} \underline{\gamma}
$$
$$
\text{Unterhalb der Schicht: } \underline{u}^{(unten)} = -\frac{1}{2} \underline{\gamma}
$$
$$
\text{Direkt auf der Schicht: } \underline{u} = \frac{1}{2} \underline{\gamma}
$$

16. Ein Wirbelring trifft auf eine Wand, die genau senkrecht zu seiner Propagationsge-
schwindigkeit steht. Was passiert

- Wirbelring erzeugt einen Spiegelwirbelring auf der anderen Seite der Wand
- Dieser Spiegelring induziert eine Geschwindigkeit, die den echten Ring von der Wand wegdrückt.
- Der Ring wird abgebremst, deformiert und bewegt sich anschließend von der Wand weg
