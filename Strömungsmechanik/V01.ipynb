2. Wie kann ein Wirbel definiert werden?
    
- Ein Wirbel ist ein Strömungsgebiet, in dem Fluidelemente eine Rotation um eine Achse aufweisen.

3. Können Wirbel nur in rotationsbehafteten Strömungen existieren? Stellen alle rotationsbehafteten Strömungen Wirbel dar? Finden Sie jeweils Beispiele, um Ihre Aussagen zu erläutern.
  
- Nein. Nicht jede rotationsbehaftete Strömung ist ein Wirbel (z. B. Couette-Strömung zwischen bewegten Platten)
- nicht jede Wirbelströmung hat eine makroskopische Rotation (z. B. Potentialwirbel)
- Beispiel: Ein Potentialwirbel besitzt eine Umlaufbewegung, aber keine lokale Rotation der Fluidelemente.


4. Wie ist die Zirkulation Γ definiert? Nutzen Sie eine Skizze zur Erläuterung.

    
- Die Zirkulation Γ ist definiert als das Linienintegral der Geschwindigkeit entlang eines geschlossenen Weges s:

$$
\Gamma = \oint_S \underline{u} \cdot d\underline{s} = \int_A \underline{\omega} \cdot \underline{n} \, dA
$$


5. Formulieren Sie in Worten, wie sich die Zirkulation berechnet.

- Die Zirkulation wird berechnet, indem man die Geschwindigkeit entlang eines geschlossenen Pfades integriert oder alternativ das Flächenintegral über die Rotation über die von diesem Pfad eingeschlossene Fläche berechnet.


6. Gegeben sei ein ebenes, laminares Geschwindigkeitsfeld mit geraden, parallelen Stromlinien. Wie groß ist die Zirkulation entlang einer Bahn mit der Form einer Acht?

- Zikulation ist null, da in einer laminaren und parallelen Strömung die Rotation
$$
\omega = \nabla \times \underline{u} = 0
$$


7. Skizzieren Sie, wie verschiedene Arten von mechanischen Spannungen auf Fluidelemente wirken. Welcher Spannungstyp führt zur Rotation von Fluidelementen?

- Normalspannungen: wirken senkrecht auf die Fläche → bewirken Volumenänderung

- Scherspannungen: wirken tangential auf die Fläche → bewirken Rotation oder Deformation

- Scherspannungen führen zur Rotation von Fluidelementen.



8. In welche Anteile kann die allgemeine Bewegung eines Fluidelements zerlegt werden? Verwenden Sie Skizzen zur Veranschaulichung.

 - Translation: Verschiebung des Fluidelements ohne Rotation oder Deformation
 - Rotation: Drehung des Fluidelements um seine eigene Achse
 - Deformation: Änderung der Form des Fluidelements durch Scherkräfte


9. Durch welchen physikalischen Effekt können Scherspannungen entstehen, die zur Rotation eines Fluidelements führen?

- Scherspannungen entstehen durch Geschwindigkeitsgradienten in der Strömung, die durch Viskosität verursacht werden. Diese Viskosität bewirkt, dass benachbarte Fluidelemente unterschiedliche Geschwindigkeiten haben, was zu Scherkräften und somit zu Rotation führt.
