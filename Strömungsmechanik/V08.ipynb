1. Welche Gleichung liefert den Ausgangspunkt bei der Herleitung der Bernoulli-Gleichung?

- Euler / Impulsgleichung:
$$
\left( \frac{\partial \underline{u}}{\partial t} + (\underline{u} \cdot \nabla) \underline{u} \right) = -\frac{1}{\rho} \nabla p + \underline{g}
$$

2. Warum ist es bei der Herleitung der Bernoulli-Gleichung zweckmäßig, für die Volumenkraft $\underline{f}$ ein Potential $G$ anzunehmen??

- weil man die Volumenkraft $\underline{f}$ als Gradient eines Potentials $G$ schreiben kann: $\underline{f} = \nabla G$
- einfacher zu integrieren

3. Unter welchen Bedingungen kann die Bernoulli-Gleichung für eine inkompressible Strömung angewendet werden? Veranschaulichen Sie sich die beiden Fälle jeweils anhand einer Skizze. Welchen Vorteil bietet somit eine Potentialströmung bei der Anwendung der Bernoulli-Gleichung?

- stationäre Strömung: $\frac{\partial \underline{u}}{\partial t} = 0$
- reibungsfrei
- Volumenkraft hat ein Potential
- entlang einer Stromlinie oder im gesamten Strömungsfeld, wenn rotaionsfrei z.B. Potentialströmung 
- Vorteil: Bernoulli gilt im gesamten Strömungsfeld bei Potentialströmung, nicht nur entlang einer Stromlinie

4. Wie lautet per Definition der Zusammenhang zwischen dem Geschwindigkeitsbetrag $q$ und dem Geschwindigkeitsvektor $\underline{u}$ und wie hängt in einer Potentialströmung der Betrag $q$ von der Potentialfunktion $\Phi$ ab?

- $q = |\underline{u}| = \sqrt {u_1^2 + u_2^2 + u_3^2}$
- $\underline{u} = \nabla \Phi \Rightarrow q = |\nabla \Phi| = \sqrt{\left(\frac{\partial \Phi}{\partial x}\right)^2 + \left(\frac{\partial \Phi}{\partial y}\right)^2 + \left(\frac{\partial \Phi}{\partial z}\right)^2}$

5. Notieren Sie die allgemeine Form der Bernoulli-Gleichung mithilfe der Potentialfunktion $\Phi$. Notieren Sie die Gleichung sowohl für instationäre als auch für stationäre Strömungen.

- instationär:
$$
\frac{\partial \Phi}{\partial t} + \frac{1}{2} ||\nabla \Phi||^2 + \frac{p}{\rho} + G = C(t) \quad C \in \mathbb{R}
$$

- stationär:
$$
\frac{1}{2} ||\nabla \Phi||^2 + \frac{p}{\rho} + G = C \quad C \in \mathbb{R}
$$


6. Welche Größe kann in einer horizontalen, ebenen Strömung mithilfe der Bernoulli-Gleichung berechnet werden, wenn die Potentialfunktion $\Phi$ gegeben ist?
Gilt für diese Größe das Superpositionsprinzip? Begründen Sie Ihre Aussage.

- horizontal: $G = 0$
- Druck kann man berechnen mit: $p = \rho \left( C - \frac{1}{2} ||\nabla \Phi||^2 \right)$
- **Superpositionsprinzip gilt nicht für den Druck** da, das Superpositionsprinzip für das Potential $\Phi$ und die Geschwindigkeiten $\underline{u}$ gilt
- Druck hängt quadratisch von der Geschwindigkeit ab: $p \sim -\frac{1}{2} \rho q^2$, also nicht lienar

7. Wie kann das Geschwindigkeitsfeld einer Strömung berechnet werden, wenn diese durch eine Stromfunktion $\Psi$ beschrieben wird?

- Aus der Stromfunktion erhält man die Geschwindigkeitskomponenten durch Ableiten:
$$
u = \frac{\partial \Psi}{\partial y} \quad , \quad v = -\frac{\partial \Psi}{\partial x}
$$


8. Welche Gleichung wird immer erfüllt, wenn sich das Geschwindigkeitsfeld $\underline{u}$ durch eine Stromfunktion $\Psi$ beschreiben lässt? Beweisen Sie Ihre Aussage.

- Kontinuitätsgleichung für inkompressible Strömungen:
$$
\nabla \cdot \underline{u} = \frac{\partial u}{\partial x} + \frac{\partial v}{\partial y} = \frac{\partial^2 \Psi}{\partial x \partial y} - \frac{\partial^2 \Psi}{\partial y \partial x} = 0 \qquad 
\rightarrow \qquad \frac{\partial^2 \Psi}{\partial x \partial y} = \frac{\partial^2 \Psi}{\partial y \partial x}
$$

9. Notieren Sie die Laplace-Gleichung für die Stromfunktion $\Psi$. Unter welchen Voraussetzungen gilt diese Gleichung? Weisen Sie Ihre Aussage nach.

$$
\nabla^2 \Psi = \frac{\partial^2 \Psi}{\partial x^2} + \frac{\partial^2 \Psi}{\partial y^2} = 0 \rightarrow \text{ gilt nur für wirbelfreie, inkompressible, ebene Strömungen}
$$

- Beweis
- Stromfunktion: $u = \frac{\partial \Psi}{\partial y},\quad v = -\frac{\partial \Psi}{\partial x}$
- Wirbelstärke: $\omega = \frac{\partial v}{\partial x} - \frac{\partial u}{\partial y}$
- Einsetzen folgt: $\omega = -\nabla^2 \Psi$
