1. Gilt das Biot–Savart-Gesetz für rotationsbehaftete Strömungen?

2. Wiederholen Sie die Definition der materiellen Ableitung und erläutern Sie dabei die einzelnen Terme.

Für Skalarefeld $\phi(\underline{x}, t)$ gilt die materielle Ableitung:
$$
\frac{D \phi}{D t} = \frac{\partial \phi}{\partial t} + \underline{u} \cdot \nabla \phi
$$

- $\frac{\partial \phi}{\partial t}$ ... Lokale zeitliche Änderung des Skalarfeldes an einem festen Ort.
- $\underline{u} \cdot \nabla \phi$ ... Konvektive Änderung des Skalarfeldes durch Bewegung des Fluidelements im Feld.


3. Notieren Sie die allgemeine Form einer Transportgleichung. Nennen Sie Beispiele für Prozesse, die durch eine solche Gleichung beschrieben werden können.

$$
\frac{D \phi}{D t} = \frac{\partial \phi}{\partial t} + \underline{u} \cdot \nabla \phi = \underbrace{\nabla \cdot (\Gamma \nabla \phi)}_{\text{Diffusion}} + \underbrace{S}_{\text{Quelle/Senke}}
$$

- $\Gamma$ ... Diffusionskoeffizient

- **Beispiele:**
  - Wärmetransport + Konvektion $\phi = T$
  - Impulstransport (Navier-Stokes-Gleichung) $\phi = \underline{u}$
  - Stofftransport + Konvektion $\phi = c$ (Konzentration eines Stoff)

4. Was sind die Annahmen und was ist der Ansatz bei der Herleitung der Wirbeltransportgleichung?

- (1) Newtonsches Fluid
  - Geschwindigkeit, Druck, Dichte sind stetig differenzierbar
  - Wirbelstärke $\underline{\omega} = \nabla \times \underline{u}$ existiert
- (2) Navier-Stokes-Gleichung gilt
- (3) Körperkräfte sind Potentialkräfte: $\underline{f} = - \nabla G$

6. Warum ist es angenehm, für die Herleitung der Wirbeltransportgleichung Potentialkräfte anzunehmen?

- Wenn Körperkräfte Potentialkräfte sind: $\underline{f} = - \nabla G$
- dann verschindet bei Navier-Stokes-Gleichung den Term $\nabla \times \underline{f} = \nabla \times (- \nabla G) = 0$

9. Wie wirken sich Reibungseffekte auf die Wirbelstärke bei der Umströmung eines stumpfen Körpers aus?
 
- Bei der Umströmung eines stumpfen Körpers erzeugt Reibung (Viskosität) starke Wirbelstärke an der Wand (No-Slip).
- Diese Wirbelstärke diffundiert ins Fluid und bildet nach der Ablösung große Wirbelstrukturen

14. Wie verändert sich die Wirbeltransportgleichung für die Annahme einer reibungsfreien, dreidimensionalen Strömung mit $\rho \neq const.$ im Vergleich zu einer reibungsbehafteten, inkompressiblen Strömung?

- Inkompressible, reibungsbehaftete (viskose) Strömung:
$$
\frac{\partial \underline{\omega}}{\partial t} + (\underline{u} \cdot \nabla) \underline{\omega} = (\underline{\omega} \cdot \nabla) \underline{u} + \nu \nabla^2 \underline{\omega}
$$

- Reibungsfreie, kompressible Strömung (variable Dichte):
$$
\frac{\partial \underline{\omega}}{\partial t} + (\underline{u} \cdot \nabla) \underline{\omega} = (\underline{\omega} \cdot \nabla) \underline{u} + \frac{1}{\rho^2} \nabla \rho \times \nabla p
$$


15.  Welcher Zusammenhang besteht zwischen dem Kelvinschen Wirbelsatz und den drei Helmholtzschen Wirbelsätzen?
- Kelvinscher Wirbelsatz: In einem idealen Fluid bleibt die Zirkulation entlang einer geschlossenen Materiallinie konstant.
$$
\frac{D \Gamma}{D t} = 0
$$

- Aus dem kelvinschen Wirbelsatz lassen sich die drei Helmholtzschen Wirbelsätze ableiten:
  1. Rotationsfreie Fluidelemente bleiben für alle Zeiten rotationsfrei.
     - $\underline{\omega}(t=t_0) = 0 \text{ zu } \underline{\omega}(t>t_0) = 0$
  2. Fluidelemente auf einer einer Wirbellinie bleiben auf dieser Wirbellinie.
  3. Intensität (Zirkulation) einer Wirbelröhre ist über alle Zeiten konstant.
     - $\Gamma(t=t_0) = \Gamma(t>t_0)$
     - $\frac{D \Gamma}{D t} = 0$

21. Warum bewegt sich ein Wirbelring in einer ansonsten ruhigen Umgebung?

Ein Wirbelring bewegt sich, weil seine eigene Wirbelstärke über das Biot–Savart-Gesetz ein Geschwindigkeitsfeld erzeugt, das an der Lage des Rings eine axiale Geschwindigkeit ergibt.
Dadurch „zieht“ sich der Ring selbst in eine Richtung voran – ganz ohne äußere Strömung.
