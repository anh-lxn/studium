1. Was ist eine Laurent-Reihe?

- Reihenentwicklung einer komplexen Funktion, die auch negative Potenzen von $(z-a)$ enthält und zur Beschreibung von Singularitäten verwendet wird.
$$
f(z) = \underbrace{\sum_{n=0}^{\infty} c_n (z-a)^n}_\text{regulärer Teil (ohne Singularitäten)}  + \underbrace{\sum_{n=1}^{\infty} \frac{c_n}{(z-a)^n}}_\text{singulärer Teil (mit Singularitäten)}
$$

- Beispiel:
$$
f(z) = \underbrace{z}_\text{regulärer Teil} + \underbrace{\frac{1}{z}}_\text{singulärer Teil}
$$


2. Entwickeln Sie die komplexe Funktion $f(z) = \frac{1}{z-z_m}$ als Laurant-Reihe um den Punkt $z_m \in \mathbb{C}$.

- kein regulärer Teil, nur singulärer Teil mit Exponenten $n=-1$ und Koeffizienten $a_{-1} = 1$:
$$
f(z) = \frac{1}{z-z_m} = (z-z_m)^{-1}
$$

3. Entwickeln Sie die komplexe Funktion $f(z) = \frac{1}{z-z_m}$ als Laurant-Reihe um den Punkt $z_0 = 0$ mit $z_0 \neq z_m$.

- Fall A: |z| < |z_m|
$$
f(z) = \frac{1}{z-z_m} = -\frac{1}{z_m} \cdot \frac{1}{1-\frac{z}{z_m}}
$$
mit der geometrischen Reihe $\frac{1}{1-q} = \sum_{n=0}^{\infty} q^n$ für |q| < 1 (hier $q = \frac{z}{z_m}$) folgt:
$$
\frac{1}{z-z_m} = -\frac{1}{z_m} \sum_{n=0}^{\infty} \left(\frac{z}{z_m}\right)^n = -\sum_{n=0}^{\infty} \frac{z^n}{z_m^{n+1}}
$$

- Fall B: |z| > |z_m|
$$
f(z) = \frac{1}{z-z_m} = \frac{1}{z} \cdot \frac{1}{1-\frac{z_m}{z}}
$$
mit der geometrischen Reihe $\frac{1}{1-q} = \sum_{n=0}^{\infty} q^n$ für |q| < 1 (hier $q = \frac{z_m}{z}$) folgt:
$$
\frac{1}{z-z_m} = \frac{1}{z} \sum_{n=0}^{\infty} \left(\frac{z_m}{z}\right)^n = \sum_{n=0}^{\infty} \frac{z_m^n}{z^{n+1}}
$$

4. Was sagt der Residuensatz aus?

- Voraussetzung: $f(z)$ ist anayltisch
- Kernaussage: Das Integral einer komplexen Funktion $f(z)$ entlang einer geschlossenen Kurve $C$ ist gleich der Summe der Residuen der Funktion innerhalb der Kurve, multipliziert mit $2 \pi i$.
$$
\oint_{C} f(z) \, dz = 2 \pi i \sum_k Res(f, a_k)
$$

5. Wie kann in der komplexen Ebene der Vektor, der durch die komplexe Zahl $z = x + iy$ beschrieben wird, um den Winkel $\phi$ in mathematisch positive Richtung gedreht werden?

- einfach multiplitzieren mit der komplexen Zahl $e^{i \phi} = \cos(\phi) + i \sin(\phi)$:
- also $z' = z \cdot e^{i \phi} = (x + iy)(\cos(\phi) + i \sin(\phi))$

6. Zeigen Sie, dass in der komplexen Ebene die durch $z$ und $iz$ beschriebenen Vektoren senkrecht aufeinander stehen.

- Betrachte die komplexen Zahlen $z = x + iy$ und $iz = i(x + iy) = -y + ix$.
- Vektoren:
  - $z$ entspricht dem Vektor $\begin{pmatrix} x \\ y \end{pmatrix}$
  - $iz$ entspricht dem Vektor $\begin{pmatrix} -y \\ x \end{pmatrix}$
- Skalarprodukt: $\begin{pmatrix} x \\ y \end{pmatrix} \cdot \begin{pmatrix} -y \\ x \end{pmatrix} = x \cdot (-y) + y \cdot x = -xy + xy = 0$

- $Q_\text{Gesamt} = 0$

8. Wie lautet die Blasius-Formel?

$$
K = i \frac{\rho}{2} \oint_C \left(\frac{dF}{dz}\right)^2 \, dz
$$

9. Wie können Sie in der Blasius-Formel den Integranden $(\frac{dF}{dz})^2$ mithilfe der komplexen Geschwindigkeitskomponente $w(z)$ ausdrücken?

- Da $w(z) = \frac{dF}{dz}$ gilt, kann der Integrand wie folgt umgeschrieben werden:
$$
\left(\frac{dF}{dz}\right)^2 = (w(z))^2
$$  

10. Erläutern Sie das schrittweise Vorgehen zur Bestimmung der Kraft auf einen Körper in einer Potentialströmung, indem Sie den Residuensatz verwenden, anstatt die Kraft manuell durch Integration entlang der Körperkontur auszurechnen.

- Komplexe Geschwindigkeit $w(z) = \frac{dF}{dz}$
- Blasius-Formel: $K = i \frac{\rho}{2} \oint_C (w(z))^2 \, dz$
- Bestimmen der Singularitäten von $w(z)$ innerhalb der Kontur $C$
- Berechnen der Residuen von $(w(z))^2$ an den Singularitäten
- Anwenden des Residuensatzes:
$$
\oint_C (w(z))^2 \, dz = 2 \pi i \sum_k Res((w(z))^2, z_k)
$$ 
- Einsetzen in die Blasius-Formel zur Berechnung der Kraft:
$$
K = i \frac{\rho}{2} \cdot 2 \pi i \sum_k Res((w(z))^2, z_k) = -\pi \rho \sum_k Res((w(z))^2, z_k)
$$

11.  Wie lautet der Kutta-Joukowskische Satz?

- Auftriebskraft $K [\frac{N}{m}]$ ist direkt Proportional zur Zirkulation $\Gamma$:
$$
K = i \rho W_\infty \Gamma
$$

- Auftrieb $A [\frac{N}{m}]$:
$$
A = -\rho W_\infty \Gamma
$$


12. Leiten Sie aus dem Kutta-Joukowskischen Satz zentrale Eigenschaften der Kraft auf einen Körper in einer Potentialströmung ab. Gehen Sie dabei u.a. auf das d’Alembertsche Paradoxon ein.

- Aus dem Kutta-Joukowskischen Satz $A = -\rho W_\infty \Gamma$ folgt:
  - Auftrieb ist proportional zur Zirkulation $\Gamma$.
  - Kraft senkrecht zur Anströmrichtung (nur Auftrieb, keine Widerstandskraft).
  - kein Widerstand in einer idealen Potentialströmung (d’Alembertsches Paradoxon).
  - Betrag der Kraft unabhängig von der Form des Körpers, solange die Zirkulation gleich bleibt.

15.  Beschreiben Sie fünf verschiedene Verfahrensweisen und Konzepte, um eine Potentialströmung zu berechnen, und bewerten Sie deren Relevanz für praktische Anwendungen.

- Analytische Lösungen:
  - Beschreibung: Verwendung von geschlossenen Formeln für einfache Geometrien (z.B. Zylinder, Platte).
  - Relevanz: Nützlich für grundlegende Studien und Validierung numerischer Methoden, aber begrenzt auf einfache Formen.
- Singularitätenmethoden:
  - Beschreibung: Der Körper wird durch eine kontinuierliche oder diskrete Verteilung von Quellen, Senken und Dipolen auf der Kontur ersetzt.
  - Relevanz: Für beliebig 2D-Körper
- Konforme Abbildungen:
  - Beschreibung: Transformation komplexer Geometrien in einfachere Formen, um analytische Lösungen anzuwenden.
  - Relevanz: Effektiv für 2D-Probleme mit bestimmten Geometrien, aber begrenzt auf spezielle Fälle.
- Numerische Lösung der Laplace-Gleichung:
  - Beschreibung: Lösung von $\nabla^2 \phi = 0$ mit FEM/FDM
  - Relevanz: Grundlage CFD für komplexe Geometrien
