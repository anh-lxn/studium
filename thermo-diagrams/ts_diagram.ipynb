```python
import matplotlib.pyplot as plt
import numpy as np
import CoolProp as CP
```

```python
fluid = "Water"

# Temperaturbereich zwischen Tripelpunkt und Kritischem Punkt
#T_min = CP.PropsSI("Ttriple", fluid)
T_min = 100 # K
print(T_min)
T_max = CP.PropsSI("Tcrit", fluid)
T = np.linspace(T_min, T_max, 500)

# Sättigungslinien
s_l = [CP.PropsSI("S", "T", Ti, "Q", 0, fluid) for Ti in T]  # Flüssig (x=0)
s_v = [CP.PropsSI("S", "T", Ti, "Q", 1, fluid) for Ti in T]  # Dampf (x=1)

# Isobaren (Beispieldrücke)
pressures = [1e5, 5e5, 1e6, 2e6, 5e6]  # Pa
for p in pressures:
    T_line = np.linspace(30<0, 700, 200)
    s_line = [CP.PropsSI("S", "T", Ti, "P", p, fluid) for Ti in T_line]
    plt.plot(s_line, T_line, label=f"{p/1e5:.1f} bar")

# Plot
plt.plot(s_l, T, "r-", lw=2, label="x=0 (Siedelinie)")
plt.plot(s_v, T, "r-", lw=2, label="x=1 (Taulinie)")
plt.xlabel("Espezifische Entropie s [J/kg*K]")
plt.ylabel("Temperatur T [K]")
plt.title("T-s-Diagramm für Wasser mit Isobaren")
plt.legend()
plt.grid(True)
plt.show()
```
