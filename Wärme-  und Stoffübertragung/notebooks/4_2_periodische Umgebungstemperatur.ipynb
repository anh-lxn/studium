```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dk = 10e-3 # m
lamdak = 14.5 # W/mK
rhok = 7900 # kg/m3
ck = 477 # J/kgK

thetaU_min = 400 # °C
thetaU_max = 800 # °C
alphaU = 100 # W/m2K
tA = 5 * 60 # s
```

```python
w = 2*np.pi / tA
ak = lamdak / (rhok * ck)
M = np.sqrt(w / (2 * ak)) * dk/2
```

```python
thetaA = thetaU_max - (thetaU_max + thetaU_min) / 2 # K
Fo = ak / ((dk/2)**2 * w)
Bi = alphaU * dk / (2 * lamdak)
m = 3

t = np.linspace(0, 2*tA, 1000)
den = 1/(m*Bi*Fo)**2 + 1
thetak = thetaA * (np.cos(w*t) + 1/(m*Bi*Fo)*np.sin(w*t)) / den
theta_umg = thetaA * np.cos(w*t) + (thetaU_max + thetaU_min) / 2
phi = 1/w * np.arctan(1/(m*Bi*Fo))

plt.plot(t, thetak + (thetaU_max + thetaU_min) / 2)
plt.plot(t, theta_umg)
plt.xlabel('Zeit t (s)')
plt.ylabel('Temperatur θk (°C)')
plt.title('Temperaturverlauf in der Kugel bei periodischer Umgebungstemperatur')
plt.legend(['Kugel', 'Umgebung'])
plt.grid()
plt.show()
phi
```
